                                         # MESH GRID
The mesh grid is very useful to build a higher dimentional grid, eg. 2D grid, such that value of multivariable function can be calculated and perhaps visualized. f(x,y). The integrand (the function[x,y]) of 2 varibale function can be imagined as 3rd axis value in 3D plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,3,6)
y = np.arange(1,2, 0.2)
```

```python
# 1-D arrays
x.shape, y.shape
```

```python
X, Y = np.meshgrid(x,y)
```

### Here, each element in the X array corresponds to the x-coordinate value of a point on the two-dimensional grid, and each element in the y array corresponds to the y-coordinate value of the same point.
Eg. If there are total 4 points on the 2D grid space. Then the X (2D) matrix will contain, x values of all the grid points, similarly for Y matrix as well. 
Let's say we have : (1,1), (1,2), (2,1), (2,2) as the four grid points then X matrix will contain: [[1,1],[2,2]] , 
whereas Y matrix: [[1,2],[1,2]]

```python
# 2-D arrays
X.shape, Y.shape
```

```python
X
```

```python
Y
```

```python
# write the function
```

```python
fun_yx = np.sin(X)*np.exp(Y)
```

```python
fun_yx.shape, fun_yx[3,1]
```

### If you compare that with the definition of our one-dimensional coordinate arrays x and y, you notice that obviously meshgrid has put all the values into axis 0 of the resulting meshed grid and the values into axis 1.
Therefore, So when you want to user the function value for the point (x[1], y[3]), then we would have to use the syntax : fun[3,1]

### In our case, we would prefer the natural ordering we are used from mathematical functions, instead. We can easily change the behavior of meshgrid by giving it the following indexing='ij' option


```python
X_, Y_ = np.meshgrid(x, y, indexing='ij')
```

```python
X_.shape, Y_.shape
```

```python
X_
```

```python
Y_
```

```python
fun_xy = np.sin(X_)*np.exp(Y_)
```

```python
fun_xy.shape, fun_xy[1,3]
```

### There's one other alternative to using the indexing in meshgrid, we can just tranport the X and Y meshgrid and get same matrices as found in X_ and Y_.

```python
X = X.T
Y = Y.T
```

```python
if X.all()==X_.all():
    print("they are same")
```

```python
integrand_xy = abs(np.sin(X)*np.tan(Y))**2
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection = "3d")

surface = ax.plot_surface(X, Y, integrand_xy, cmap = plt.cm.cividis)

ax.set_xlabel("X", labelpad = 20)
ax.set_ylabel("Y", labelpad = 20)
ax.set_zlabel("Integrand", labelpad =20)
fig.colorbar(surface, shrink=0.5, aspect=8)
plt.show()
```
